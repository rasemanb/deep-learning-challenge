## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
df = application_df.drop(['EIN'], axis=1)
df.head()
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE
# Keeping values above 50 rather than 200
apps_to_replace = ['T25', 'T14', 'T29', 'T15', 'T17', 'T2', 'T12']
# Replace in dataframe
for app in apps_to_replace:
   df['APPLICATION_TYPE'] = df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
counts = df['CLASSIFICATION'].value_counts(dropna=False) 
counts[counts>3]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
# Keeping values above 200

class_to_replace = ['C4000', 
                    'C5000', 
                    'C1270', 
                    'C2800', 
                    'C7100', 
                    'C1300', 
                    'C1280', 
                    'C1230', 
                    'C1400', 
                    'C7200', 
                    'C2300', 
                    'C1240', 
                    'C8000', 
                    'C7120', 
                    'C1500', 
                    'C1800', 
                    'C6000', 
                    'C1250', 
                    'C8200', 
                    'C1238', 
                    'C1278', 
                    'C1235', 
                    'C1237', 
                    'C7210', 
                    'C2400', 
                    'C1720', 
                    'C4100', 
                    'C1257', 
                    'C1600', 
                    'C2710', 
                    'C0', 
                    'C1260', 
                    'C3200', 
                    'C1246', 
                    'C1234', 
                    'C1256', 
                    'C1267', 
                    'C1728', 
                    'C1732', 
                    'C1900', 
                    'C1820', 
                    'C6100', 
                    'C1248', 
                    'C2600', 
                    'C1370', 
                    'C5200', 
                    'C2561',
                    'C1245',
                    'C4200',
                    'C2190',
                    'C4500',
                    'C8210',
                    'C1236',
                    'C4120',
                    'C2570',
                    'C1580',
                    'C3700',
                    'C2170',
                    'C2500',
                    'C1570',
                    'C1283',
                    'C2380',
                    'C2150',
                    'C2700']

# Replace in dataframe
for cls in class_to_replace:
    df['CLASSIFICATION'] = df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
df['CLASSIFICATION'].value_counts()
```

```python
# Cutting down NAME column to save RAM and runtime
#  YOUR CODE GOES HERE
ct = df['NAME'].value_counts(dropna=False) 
ct[ct>10]
```

```python
names_to_replace = list(ct[ct<10].index)

# Replace in dataframe
for name in names_to_replace:
    df['NAME'] = df['NAME'].replace(name,"Other")
    
# Check to make sure binning was successful
df['NAME'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
encoded = pd.get_dummies(df, dtype=float)

encoded.head()
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = encoded['IS_SUCCESSFUL']
X = encoded.drop('IS_SUCCESSFUL', 1).values
# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=15)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
num_input = len(X_train[0])
hl1 = 9
hl2 = 6
hl3 = 6

nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hl1, input_dim = num_input, activation='relu'))

# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hl2, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
#  YOUR CODE GOES HERE
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
